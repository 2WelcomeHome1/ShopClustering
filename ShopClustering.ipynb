***Импортируем библиотеки***

```python
import pandas as pd # Для работы с табличными данными
import numpy as np # Для работы с массивами

import matplotlib # Для отрисовки графиков
import matplotlib.pyplot as plt # Для отрисовки графиков

# выбор темы картинок (необязательная строчка)
matplotlib.style.use('ggplot')
# Эта строчка нужна для того, чтобы картинки отображались в ячейках
%matplotlib inline
```

***Загружаем данные***

```python
df_branches = pd.read_csv('branch.dat', sep = ';',index_col = 'НАЗВАНИЕ',  encoding = 'cp1251')
df_branches
```

***Выполняем преобразование данных***

Консультатнт - бинарная переменная

```python
df_branches['КОНСУЛЬТ'] = df_branches['КОНСУЛЬТ'].map({'Есть': 1, 'Нет ': 0})
df_branches
```

КОНКУРЕН - преобразуем в ранговые данные от 0 до 1, где 1 - это наилучшее значение, а 0 - наихудшее

```python
df_branches['КОНКУРЕН'].unique()
```

```python
df_branches['КОНКУРЕН'] = df_branches['КОНКУРЕН'].map({'лучше ': 1, 'одинак': 0.5, 'хуже  ': 0 })
df_branches
```

АССОРТИМ - Повторяем раннее подобранное преобразование, но условно примем, что широкий ассортимент приравнивается максимальному, или приблизительно равен ему. 

```python
df_branches['АССОРТИМ'].unique()
```

```python
df_branches['АССОРТИМ'] = df_branches['АССОРТИМ'].map({'широкий': 1,'макс   ': 1, 'средний': 0.5, 'миним  ': 0 })
df_branches
```

Далее выберем преобразование для Дизайна с помощью процедуры One Hot Encoding, где каждое обознаение является категориальной переменной, и может либо как существовать, либо не быть вообще

```python
df_branches['ДИЗАЙН'].unique()
```

```python
df_branches = pd.get_dummies(df_branches, columns=["ДИЗАЙН"])
df_branches
```

Удалем столбец продажи, дополнительно сохраняем его в отдельную серию

```python
Sales = df_branches['ПРОДАЖИ']
df_branches = df_branches.drop('ПРОДАЖИ' , axis = 1)
```

```python
df_branches
```

***Нормализация данных***

Выполняем преобразование данных. Переводим абсолютное значение каждого элемента в тренировочном наборе к диапзону от 1 до 0, используя логику из предыдущих примеров. Благодвря данном методу нормализации максимальное значение столбца автоматически приравнивается к еденице, а минимальное к нулю

```python
from sklearn import preprocessing

norm = preprocessing.MaxAbsScaler()
norm.fit(df_branches)

df_branches_norm = norm.transform(df_branches)
df_branches_norm = pd.DataFrame(df_branches_norm, index = df_branches.index, columns = df_branches.columns)
df_branches_norm
```

***Cтроим дендограмму***

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

link = linkage(df_branches_norm, 'ward', 'euclidean')

dn_branches_norm = dendrogram(link,
                # массив с названиями объектов для подписи по оси X
                labels = df_branches_norm.index.tolist(),
                # повернуть подписи вертикально
                leaf_rotation = 90)
```
